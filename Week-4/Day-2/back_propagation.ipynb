```python
def initialize_parameters(layers_dims):
    np.random.seed(3)
    parameters={}
    L=len(layers_dims)
    for l in range(1,L):
        parameters['W'+str(l)]=np.ones((layers_dims[l-1],layers_dims[1]))*0.1
        parameters['b'+str(l)]=np.zeros((layers_dims[1],1))
        
    return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
def linear_forward(A_prev,W,b):

    Z=np.dot(W.T,A_prev)+b
    return Z
```

```python
def L_layer_forward(X,parameters):
    A=X
    L=len(parameters)//2

    for l in range(1,L+1):
        A_prev=A
        W1=parameters["W",str(l)]
        b1=parameters["b"+str(l)]
        print("A"+str(l-1)+": ",A_prev)
        print("W"+str(l)+": ",W1)
        print("b"+str(l)+": ",b1)
        print("--"*20)

        A=linear_forward(A_prev,W1,b1)
        print("A"+str(l)+": ",A)
        print("**"*20)

    return A,A_prev
```
